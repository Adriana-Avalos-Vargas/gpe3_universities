```python
## Dependencies
import requests # to make api requests
import json #To manage api respose as json file
import time #To add sleep time in a computational process
import pandas as pd

# Google developer API key
from config import gkey
```

```python
#Lets call the df with weather info
universities_path = "data/universities_ranking.csv"
```

```python
universities_df = pd.read_csv(universities_path)
universities_df.head()
```

```python
#list of universities
universities = universities_df["title]
universities
```

```python
##Lets start the query for the vactions_df
#To do so, first we need to add new columns to vactions_df, where info is going to be stores
# Add columns for lat, lng, airport name, airport address, airport rating
# Note that we used "" to specify initial entry.
universities_df["Latitude"] = ""
universities_df["Longitude"] = ""
universities_df.head()
```

```python
#
# Target city
target_city = "University of Oxford"

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, gkey)

geo_data = requests.get(target_url).json()

# Print the json
print(geo_data)
```

```python
lat = geo_data["results"][0]["geometry"]["location"]["lat"]
lng = geo_data["results"][0]["geometry"]["location"]["lng"]
```

```python
#Lets make the retreive
# Lets define the query
#Create a counter for the times that the API did find a city
number_cities = 0

#parameters dictionary to update each iteration the keyword lodgind is added to ensure it is a hotel and not another kind of 
#bussines that might have in its name the world hotel
# Use the lat/lng we recovered to search for hotels
for index, row in universities_df.iterrows():
    # get the name of the university
    target_university = row["title"]

    # Use the search term: "International Airport" and our lat/lng
    base_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_university, gkey)

    # make request
    name_address = requests.get(base_url)
    

    # convert to json
    name_address = name_address.json()
    

    # Since some data may be missing we incorporate a try-except to skip any that are missing a data point.
    try:
        universities_df.loc[index, "Latitude"] = name_address["results"][0]["geometry"]["location"]["lat"]
        universities_df.loc[index, "Longitude"] = name_address["results"][0]["geometry"]["location"]["lng"]
        number_cities += 1
        #Lets print a mesage
        print(f"Information of {target_university} retrieved, call number {number_cities}")
        time.sleep(1)
    except (KeyError, IndexError):
        number_cities += 1
        print(f"Missing field/result... skipping call number {number_cities}")
```
